# Data Munging/DATA wrangling /cleaning the data


```python
#Check missing values in the dataset
data.isnull()  
#null values is true otherwise False
data.apply(lambda x: sum(x.isnull()),axis=0) 
```

```python
data.isnull().info()

#print number of person  have Null(nan) value in each column 
data.isnull().sum()
```

```python
#note:::missing values may not be always Nans.if Loan amoutn term is 0 then there is no meaning or no sense 
```

```python
#fill LoanAmount missing values 
data['LoanAmount'].fillna(data['LoanAmount'].mean(), inplace=True)
```

```python
#sel_employed have missing value  
data['Self_Employed'].value_counts()   #frequency table 
#since ~86% values are No,it is safe to frame missing values as "NO" as high probabilty of success 

#fill missing value in self_employed columns
data["Self_Employed"].fillna("No",inplace=True)
data.isnull().sum()
```

```python
#create a pivot table for finding median values for all the groups of unique values of self_employed and education features 
table=data.pivot_table(index="Self_Employed",columns="Education",aggfunc=np.median)
table
```

```python
def fage(x):
    return table.loc[x['Self_Employed'],x['Education']] 
data['LoanAmount'].fillna(data[data['LoanAmount'].isnull()].apply(fage, axis=1), inplace=True)
```

```python
#some peoples migt apply for high value loans due to specific needs .so instead of treeating them outliers,lets try alog transformation to nullify their effect 
data['LoanAmount_log'] = np.log(data['LoanAmount']) 
data['LoanAmount_log'].hist(bins=20)
```
